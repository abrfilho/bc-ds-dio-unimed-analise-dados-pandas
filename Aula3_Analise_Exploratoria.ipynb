```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-darkgrid")
```

```python
df = pd.read_excel("datasets\AdventureWorks.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
#Receita total
round(df["Valor Venda"].sum(), 2)
```

```python
#Custo total
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
round(df["Custo"].sum(), 2)
```

```python
#Lucro total
df["Lucro"] = df["Valor Venda"] - df["Custo"]
round(df["Lucro"].sum(), 2)
```

```python
#Tempo de envio (intervalor entre a compra e envio)
df["Tempo de envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.groupby("Marca")["Tempo de envio"].mean()
```

```python
#Verificar a presença de valores nulos
df.isnull().sum()
```

```python
#Agrupamento por ano e marca
pd.options.display.float_format="{:20,.2f}".format
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total de produtos vendidos", xlabel="Total", ylabel="Produto");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano", xlabel="Ano", ylabel="Lucro");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro x Mês", xlabel="Mês", ylabel="Lucro");
```

```python
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro x Marca", xlabel="Marca", ylabel="Lucro");
plt.xticks(rotation="horizontal")
```

```python
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro x Classe", xlabel="Classe", ylabel="Lucro");
plt.xticks(rotation="horizontal")
```
